```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(os.path.join('Trinity Student Insights Survey 2026(1-184).csv'))
```

```python
df['On average, how many hours per day do you spend on your phone?'].value_counts()
```

```python
df['What grade do you usually achieve overall?'] = df['What grade do you usually achieve overall?'].replace(['A+', 'A', 'B+', 'B', 'C', 'D or below'], [0, 1, 2, 3, 4, 5], inplace=True)
#df['On average, how many hours per day do you spend on your phone?'] = df['On average, how many hours per day do you spend on your phone?'].replace(['0–2 hours', '3–4 hours', '5–6 hours', '7–8 hours', '9+ hours'], [0, 1, 2, 3, 4], inplace=True)
df["How long does it take you to travel to school (one way)?"] = df["How long does it take you to travel to school (one way)?"].replace(["Under 15 minutes", "15–30 minutes", "31–60 minutes", "More than 60 minutes"], [0, 1, 2, 3], inplace=True)
```

```python
df['What grade do you usually achieve overall?'].value_counts()
```

```python
ig3, ax3 = plt.subplots(figsize=(8, 5))
survival_counts = pd.crosstab(df["What grade do you usually achieve overall?"], df["On average, how many hours per day do you spend on your phone?"])
survival_counts.plot(kind='bar', stacked=True, color=['lightcoral', 'lemonchiffon', 'lightgreen', 'lightblue', 'lightsalmon'], ax=ax3)
ax3.set_title('How many hours different average grades spend on their phones', fontsize=16)
ax3.set_xlabel('Average grade', fontsize=12)
ax3.set_ylabel('Count', fontsize=12)
ax3.legend(['0-2', '3-4', '5-6', '7-8', '9+'], title='Screen time')
ax3.tick_params(axis='x', rotation=0)
```

```python
corr = df['On average, how many hours per day do you spend on your phone?'].corr(df['What grade do you usually achieve overall?'])
print(corr)
```

```python
df['On average, how many hours per day do you spend on your phone?'] = df['On average, how many hours per day do you spend on your phone?'].replace(['0–2 hours', '3–4 hours', '5–6 hours', '7–8 hours', '9+ hours'], [1, 3.5, 5.5, 7.5, 9], inplace=True)
```

```python
print(df['On average, how many hours per day do you spend on your phone?'].mean())
df.groupby('What grade do you usually achieve overall?')['On average, how many hours per day do you spend on your phone?'].mean()
```

```python
df['How many days per week do you participate in physical activity?'] = df['How many days per week do you participate in physical activity?'].replace(['0', '1-2', '3-4', '5-6', '7'], [0, 1, 3.5, 5.5, 7], inplace=True)
df['Dogs or cats?'] = df['Dogs or cats?'].replace(['Dogs', 'Cats'], [0, 1], inplace=True)
```

```python
corr = df['Dogs or cats?'].corr(df['How many days per week do you participate in physical activity?'])
print(corr)
```

```python
ig3, ax3 = plt.subplots(figsize=(8, 5))
survival_counts = pd.crosstab(df['How many days per week do you participate in physical activity?'], df["Dogs or cats?"])
survival_counts.plot(kind='bar', stacked=True, color=['orangered', 'Chartreuse'], ax=ax3)
ax3.set_title('Prefered pet based on days spent playing sport', fontsize=16)
ax3.set_xlabel('How many days people partake in physical activity', fontsize=12)
ax3.set_ylabel('Count', fontsize=12)
ax3.legend(['Dogs', 'Cats'], title='Dogs or Cats')
ax3.tick_params(axis='x', rotation=0)
```
