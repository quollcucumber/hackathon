```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# convert school travel times

# Less than 15 = 0
# 15-30 = 1
# 31-60 = 2
# More than 60 = 3

df = pd.read_csv("Trinity Student Insights Survey 2026(1-184).csv")

fig2, ax2 = plt.subplots(figsize=(8, 5))
counts = df["How long does it take you to travel to school (one way)?"].value_counts(sort=False)
counts.plot(kind='bar', color=['skyblue', 'orange', 'red', 'green'], ax=ax2)
ax2.set_title('Number of students grouped by travel time to school', fontsize=16)
ax2.set_xlabel('Travel time to school', fontsize=12)
ax2.set_ylabel('Number of Students', fontsize=12)
ax2.tick_params(axis='x', rotation=0)
```

```python
df["How long does it take you to travel to school (one way)?"] = df["How long does it take you to travel to school (one way)?"].replace(["Under 15 minutes", "15–30 minutes", "31–60 minutes", "More than 60 minutes"], [0, 1, 2, 3], inplace=True)
df['What grade do you usually achieve overall?'] = df['What grade do you usually achieve overall?'].replace(['A+', 'A', 'B', 'C', 'D or below'], [0, 1, 2, 3, 4], inplace=True)
df["How often do you use AI to help with homework?"] = df["How often do you use AI to help with homework?"].replace(["Never", "Rarely", "Sometimes", "Often", "Very often"], [0, 1, 2, 3, 4])
df["Which type of learning activities do you tend to enjoy most?"] = df["Which type of learning activities do you tend to enjoy most?"].replace(["Physical activity and movement", "Problem-solving and calculations", "Technology and coding", "Business and real-world applications", "Experiments and investigations","Reading and writing"], [0,1,2,3,4,5])
df["Which language are you currently studying (or have studied the longest)?"] = df["Which language are you currently studying (or have studied the longest)?"].replace(["Latin", "French", "German", "Chinese", "Not studying a language"], [0,1,2,3,4])
df["When do you usually complete your homework?"] = df["When do you usually complete your homework?"].replace(["Before school", "During school", "After school", "Late at night"], [0,1,2,3,])


print(df["How often do you use AI to help with homework?"].corr(df["What grade do you usually achieve overall?"]))
print(df["How long does it take you to travel to school (one way)?"].corr(df["What grade do you usually achieve overall?"]))
df = pd.read_csv("Trinity Student Insights Survey 2026(1-184).csv")
survival_counts = df.groupby(['Which language are you currently studying (or have studied the longest)?', 'What grade do you usually achieve overall?']).size().unstack()

# Plot a bar chart for survival by gender
fig3, ax3 = plt.subplots(figsize=(8, 5))
survival_counts.plot(kind='bar', stacked=True, ax=ax3)
ax3.set_title('Average Grade per School Language', fontsize=16)
ax3.set_xlabel('Language', fontsize=12)
ax3.set_ylabel('Students', fontsize=12)
ax3.legend(["A+", "A", "B", "C", "D or below"], title='Grades')
ax3.tick_params(axis='x', rotation=0)
```
